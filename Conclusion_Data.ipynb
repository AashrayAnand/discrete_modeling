```python
import pandas as pd

quantile_income = pd.read_csv('./monte_carlo_inputs/quantile_income.csv')
```

```python
#format the data
quantile_income = quantile_income.replace('Computer and Information Sciences and Support Services','cs')
quantile_income = quantile_income.replace('Engineering Technologies and Engineering-Related Fields','eng')
quantile_income = quantile_income.replace('Health Professions and Related Programs','heal')
quantile_income = quantile_income.replace('Bachelor\'s','Bachelors')
# majors: [cs, eng, heal, NaN]
# awd_lvl: [Associate, Bachelor's]
quantile_income['75pct'] =(quantile_income['75pct'].str[1:-1]).astype(float)*1000
quantile_income['25pct'] =(quantile_income['25pct'].str[1:-1]).astype(float)*1000

quantile_income = quantile_income.rename(index=str, columns={"awd_lvl": "education"})
bachelors = quantile_income[quantile_income["education"] == "Bachelors"]
associates = quantile_income[quantile_income["education"] == "Associate"]
dropouts = quantile_income[quantile_income["education"] == "Some College"]
```

```python
import matplotlib.pyplot as plt
```

```python
for major in set(quantile_income["major"]):
    cur_bachelors = bachelors[bachelors["major"] == major]
    cur_associates = associates[associates["major"] == major]
    cur_dropouts = dropouts[dropouts["major"] == major]
    if len(cur_bachelors) > 0 and len(cur_associates) > 0:
        normalized_bachelors_salary = list(cur_salary - 18000 for cur_salary in cur_bachelors["25pct"])
        line1 = plt.plot(cur_bachelors["years_after_grad"], normalized_bachelors_salary, 'bo-', label="bachelors")
        first_legend = plt.legend(handles=line1, loc=4)
        ax = plt.gca().add_artist(first_legend)
        normalized_associates_salary = list(cur_salary - 9000 for cur_salary in cur_associates["25pct"])
        line2 = plt.plot(cur_associates["years_after_grad"], normalized_associates_salary, 'ro-', label="associates")
        second_legend = plt.legend(handles=line2, loc=2)
        plt.xlabel("years after graduation")
        plt.ylabel("salary")
        plt.title("Salary n years after graduation for {} majors".format(major))
        plt.show()
```

```python
for major in set(quantile_income["major"]):
    cur_bachelors = bachelors[bachelors["major"] == major]
    cur_associates = associates[associates["major"] == major]
    cur_dropouts = dropouts[dropouts["major"] == major]
    if len(cur_bachelors) > 0 and len(cur_associates) > 0:
        normalized_bachelors_salary = list((cur_salary - 18000)/18000 for cur_salary in cur_bachelors["25pct"])
        line1 = plt.plot(cur_bachelors["years_after_grad"], normalized_bachelors_salary, 'bo-', label="bachelors")
        first_legend = plt.legend(handles=line1, loc=4)
        ax = plt.gca().add_artist(first_legend)
        normalized_associates_salary = list((cur_salary - 9000)/9000 for cur_salary in cur_associates["25pct"])
        line2 = plt.plot(cur_associates["years_after_grad"], normalized_associates_salary, 'ro-', label="associates")
        second_legend = plt.legend(handles=line2, loc=2)
        plt.xlabel("years after graduation")
        plt.ylabel("salary")
        plt.title("Annual relative ROI for {} majors".format(major))
        plt.show()
```
